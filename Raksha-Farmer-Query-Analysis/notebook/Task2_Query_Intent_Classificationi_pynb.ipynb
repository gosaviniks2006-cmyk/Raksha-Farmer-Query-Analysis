```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/raksha-farmer-query.csv')

# Display the first few rows to understand the data structure
print("Original DataFrame head:")
display(df.head())
```

Now, let's define the keywords for each category and create a function to classify the queries.

```python
def classify_query(query_text):
    query_text = str(query_text).lower()

    if any(keyword in query_text for keyword in ['rain', 'weather', 'storm', 'drought', 'cyclone', 'temperature', 'humidity']):
        return 'Weather'
    elif any(keyword in query_text for keyword in ['pest', 'insect', 'disease', 'fungus', 'bug', 'aphid', 'control']):
        return 'Pest'
    elif any(keyword in query_text for keyword in ['nutrient', 'fertilizer', 'soil', 'deficiency', 'urea', 'potash', 'phosphorus']):
        return 'Nutrient'
    elif any(keyword in query_text for keyword in ['market', 'price', 'sell', 'buy', 'demand', 'supply', 'trade', 'export']):
        return 'Market'
    elif any(keyword in query_text for keyword in ['scheme', 'subsidy', 'government', 'loan', 'grant', 'policy', 'pm kisan']):
        return 'Government Scheme'
    else:
        return 'Other'

# Apply the classification function to the 'QueryText' column
df['Predicted_Category'] = df['QueryText'].apply(classify_query)

# Display the DataFrame with the new 'Predicted_Category' column
print("\nDataFrame with classified queries:")
display(df.head(10))
```

Let's see the distribution of the classified categories.

```python
# Get the value counts of the 'Predicted_Category' column
category_counts = df['Predicted_Category'].value_counts()

# Display the counts
print("\nDistribution of Query Categories:")
print(category_counts)

# Optionally, display as a percentage
print("\nPercentage Distribution of Query Categories:")
print(df['Predicted_Category'].value_counts(normalize=True) * 100)
```
